1. What is Python, and why is it popular?
    - Python is a high-level, interpreted programming language known for its simple syntax and readability. It is popular because it is easy to learn, versatile, has a large community, and is widely used in web development, data science, AI, and automation.

2. What is an interpreter in Python?
    - An interpreter executes Python code line by line, converting it into machine-readable instructions at runtime.

3. What are pre-defined keywords in Python?
    - Keywords are reserved words in Python that have special meanings, such as if, else, while, for, True, False, etc.

4. Can keywords be used as variable names?
    - No, keywords cannot be used as variable names because they are reserved for specific purposes.

5. What is mutability in Python?
    - Mutability refers to whether an object’s value can be changed after it is created.

6. Why are lists mutable, but tuples are immutable?
    - Lists are designed to allow modification (adding, removing, changing elements), while tuples are fixed and cannot be changed after creation.

7. Difference between '==' and 'is'?
    - '==' checks value equality, while 'is' checks whether both variables refer to the same object in memory.

8. Logical operators in Python?
    - Logical operators are and, or, and not.

9. What is type casting?
    - Type casting means converting one data type into another.

10. Implicit vs Explicit type casting?
    - Implicit is automatic conversion by Python; explicit is manual using functions like int(), float(), str().

11. Purpose of conditional statements?
    - They allow decision-making in code based on conditions.

12. How does elif work?
    - It checks another condition if the previous if condition is false.

13. Difference between for and while loop?
    - For loop is used for known iterations; while loop is used when the condition determines repetition.

14. Scenario for while loop?
    - When the number of iterations is unknown, such as user input validation.
