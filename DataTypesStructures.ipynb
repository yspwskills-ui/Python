1. What are data structures?
    - Data structures are ways of organizing and storing data efficiently for operations like access, modification, and storage.
2. Mutable vs Immutable
    - Mutable types (list, dict) can change after creation. Immutable types (tuple, string) cannot be changed.
3. List vs Tuple
    - Lists are mutable, tuples are immutable. Lists use [], tuples use ().
4. Dictionaries storage
    - Dictionaries store data as key-value pairs using hash tables.
5. Set vs List
    - Sets remove duplicates and are faster for membership checks.
6. String vs List
    - Strings are immutable sequences of characters; lists are mutable collections.
7. Tuples integrity
    - Tuples prevent accidental changes, ensuring fixed data.
8. Hash table
    - A hash table maps keys to values quickly; dictionaries use this concept.
9. Can lists contain different types?
    - Yes, lists can hold multiple data types.
10. Why strings immutable?
    - To ensure memory efficiency and security.
11. Advantages of dict
    - Fast lookup using keys instead of index.
12. Tuple use case
    - When data should not change, like coordinates.
13. Sets duplicates
    - Sets automatically remove duplicate values.
14. 'in' keyword
    - In lists checks values; in dict checks keys.
15. Modify tuple?
    - No, tuples are immutable.
16. Nested dictionary
    - A dictionary inside another dictionary.
17. Time complexity dict access
    - O(1) average time.
18. Lists preferred
    - When order matters and duplicates allowed.
19. Dict unordered
    - Earlier unordered; now insertion-ordered but accessed via keys.
20. List vs Dict retrieval
    - Lists use index; dicts use keys.
