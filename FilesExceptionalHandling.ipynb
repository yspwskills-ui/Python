1. Difference between interpreted and compiled languages
    - Interpreted (Python): Code runs line-by-line using an interpreter. Slower but easier to debug.
    - Compiled (C, C++): Code is converted to machine code before execution. Faster but needs compilation.

2. What is exception handling in Python
    - It is a way to handle runtime errors so that the program does not crash and continues execution.

3. Purpose of the finally block
    - The finally block always executes, whether an exception occurs or not. It is mainly used for cleanup (e.g., closing files).

4. What is logging in Python
    - Logging is a way to track events, errors, and program flow using the logging module.

5. Significance of __del__ method
    - It is a destructor method that is called when an object is about to be destroyed, used for cleanup.

6. Difference between import and from ... import
    - import module → access using module.name
    - from module import name → access directly as name

7. Handling multiple exceptions
    - You can use multiple except blocks or a tuple:
except (TypeError, ValueError):

8. Purpose of with statement (files)
    - It automatically handles file opening and closing, ensuring proper resource management.

9. Difference between multithreading and multiprocessing
    - Multithreading: Multiple threads in same process (shared memory, lightweight)
    - Multiprocessing: Separate processes (more powerful, uses multiple CPUs)

10. Advantages of logging
    - Tracks errors and events
    - Helps debugging
    - Maintains records
    - Better than print statements for large programs

11. What is memory management in Python
    - It refers to how Python handles allocation and deallocation of memory automatically.

12. Basic steps in exception handling
    - try → code that may cause error
    - except → handle error
    - else → runs if no error
    - finally → always runs

13. Why memory management is important
    - It ensures efficient use of memory and prevents memory leaks.

14. Role of try and except
    - try: wraps risky code
    - except: catches and handles errors

15. Python garbage collection
    - Python automatically deletes unused objects using reference counting and garbage collector.

16. Purpose of else block
    - Executes only if no exception occurs in the try block.

17. Common logging levels
    - DEBUG
    - INFO
    - WARNING
    - ERROR
    - CRITICAL

18. Difference: os.fork() vs multiprocessing
    - os.fork() creates a child process (Unix only)
    - multiprocessing is cross-platform and easier to use

19. Importance of closing a file
    - Prevents memory leaks and data corruption, and frees system resources.

20. file.read() vs file.readline()
    - read() → reads entire file
    - readline() → reads one line at a time

21. Logging module usage
  - Used to record program events, errors, and debug information.

22. os module in file handling
    - Used for file and directory operations like create, delete, rename, etc.

23. Challenges in memory management
    - Memory leaks
    - Fragmentation
    - High memory usage

24. Raising exception manually
    - raise ValueError("Invalid input")

25. Importance of multithreading
    - Improves performance in I/O-bound tasks like file handling, networking, etc.


1. Open a file for writing and write a string
```
with open("file.txt", "w") as f:
    f.write("Hello, World!")
```

2. Read file and print each line
```
with open("file.txt", "r") as f:
    for line in f:
        print(line.strip())
```

3. Handle file not found error
```
try:
    with open("file.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File does not exist")
```

4. Read from one file and write to another
```
with open("source.txt", "r") as src:
    with open("dest.txt", "w") as dest:
        dest.write(src.read())
```

5. Handle division by zero
```
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

6. Log error on division by zero
```
import logging

logging.basicConfig(filename="app.log", level=logging.ERROR)

try:
    x = 10 / 0
except ZeroDivisionError:
    logging.error("Division by zero occurred")
```

7. Log at different levels
```
import logging

logging.basicConfig(level=logging.DEBUG)

logging.info("This is info")
logging.warning("This is warning")
logging.error("This is error")
```

8. Handle file opening error
```
try:
    f = open("file.txt", "r")
except Exception as e:
    print("Error:", e)
```

9. Read file line by line into list
```
with open("file.txt", "r") as f:
    lines = f.readlines()

print(lines)
```

10. Append data to a file
```
with open("file.txt", "a") as f:
    f.write("New line\n")
```

11. Handle missing dictionary key
```
d = {"a": 1}

try:
    print(d["b"])
except KeyError:
    print("Key not found")
```

12. Multiple except blocks
```
try:
    x = int("abc")
except ValueError:
    print("Value Error")
except TypeError:
    print("Type Error")
```

13. Check if file exists
```
import os

if os.path.exists("file.txt"):
    print("File exists")
else:
    print("File not found")
```

14. Logging info and error
```
import logging

logging.basicConfig(filename="app.log", level=logging.INFO)

logging.info("Program started")

try:
    10/0
except:
    logging.error("Error occurred")
```

15. Print file content and handle empty file
```
with open("file.txt", "r") as f:
    content = f.read()
    if content:
        print(content)
    else:
        print("File is empty")
```

16. Memory profiling (basic example)
```
from memory_profiler import profile

@profile
def my_func():
    a = [i for i in range(10000)]
    return a

my_func()
```

17. Write list of numbers to file
```
numbers = [1, 2, 3, 4, 5]

with open("numbers.txt", "w") as f:
    for num in numbers:
        f.write(str(num) + "\n")
```

18. Logging with rotation (1MB)
```
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("app.log", maxBytes=1_000_000, backupCount=3)

logging.basicConfig(handlers=[handler], level=logging.INFO)
logging.info("Logging with rotation")
```

19. Handle IndexError and KeyError
```
try:
    lst = [1, 2]
    print(lst[5])
    d = {"a": 1}
    print(d["b"])
except IndexError:
    print("Index error")
except KeyError:
    print("Key error")
```

20. Read file using context manager
```
with open("file.txt", "r") as f:
    content = f.read()
    print(content)
```

21. Count occurrences of a word
```
word = "python"

with open("file.txt", "r") as f:
    text = f.read()

print(text.lower().count(word))
```

22. Check if file is empty
```
import os

if os.path.getsize("file.txt") == 0:
    print("File is empty")
```

23. Log error during file handling
```
import logging

logging.basicConfig(filename="app.log", level=logging.ERROR)

try:
    with open("file.txt", "r") as f:
        print(f.read())
except Exception as e:
    logging.error(f"File error: {e}")
```
